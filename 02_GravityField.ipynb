**Gravity Field Project**

This script is mainly meant to model the gravitational acceleration at a distance r from the center of a star in stars with constant density, and in stars with linearly decreasing density. However, we also plot the ratio of the radial mass over the total mass of stars with constant density, and the ratio of the density over central density of stars with linearly decreasing density. In all plots, we have the ratio of the radial distance from the center of the star versus the radius of the star on the x-axis.

$$



**Constant Density Model**

The first part of the code regards a star with constant density $\rho$. First, we make a plot of the ratio of the enclosed mass $M_r$ and of the stellar mass $M_\star$ as a function of the ratio of the radial distance and of the stellar radius $\frac{r}{R_\star}$. Next, we make a plot of ratio of the gravitational acceleration $g$ at radial position $r$, and of the gravitational acceleration $g_\star$cat the surface of a star.

The enclosed mass can be calculated from the density as such:

$$M_r = \int_{0}^{2\pi} \int_{0}^{\pi} \int_{0}^{r} \rho r^2 sin(\theta) dr d\theta d\phi = \frac{4\pi}{3} r^3$$


Clearly then, the ratio of the enclosed mass at any radius, and of the mass if the star is:

$$\frac{M_r}{M_\star} = \frac{r^3}{R_\star^3} = \left(\frac{r}{R_\star}\right)^3$$

Where $R$ is the radius of the star. From elemetary physics:

$$g(r) = \frac{G M_r}{r^2}$$

And so, the ratio of the gravitational acceleration at the radius r and of the acceleration at the surface of the star is:

$$\frac{g}{g_\star} = \frac{M_r}{M_\star} \left(\frac{R_\star}{r}\right)^2$$

Our first plot will show both the value of the ratio of the masses $\frac{M_r}{M_\star}$ as a function of the ratio of the radii $\frac{r}{R_\star}$, and the value of the gravitational acceleration as a function of the aforementioned ratio.

**Linear Density Model**

The next part of the code regards a star with linearly decreasing density: 

$$\rho = \rho_c \left[1 - (r/R_\star) \right]$$

Where $\rho_c$ is the density at $r=0$. 

Next, we make a plot of the ratio of the gravitational acceleration $g$ at a radial distance $r$ from the center, and of the gravitational acceleration $g_\star$ at the surface of the star. For that purpose, we neeed to have the ratio enclosed mass $M_r$ and of the stellar mass $M_\star$ as a function of $\frac{r}{R_\star}$. 

Next, we make a plot of the ratio of the density $\rho$ at a radial distance $r$ and of the density $\rho_c$ at $r=0$ as a function of the radial ratio $\frac{r}{R_\star}$.

Finally, we find the value of the maximum density, and the value of $\frac{r}{R_\star}$ at which said maximum occurs.

The enclosed mass can be calculated from the density as such:

$$M_r = \int_{0}^{2\pi} \int_{0}^{\pi} \int_{0}^{r} \rho r^2 sin(\theta) dr d\theta d\phi = 4\pi \int_{0}^{r} \rho_c \left[1 - (r/R_\star)  \right] r^2 dr$$

 $$= 4\pi\int_{0}^{r} \rho_c \left[r^2 - (r^3/R_\star)  \right] dr = 4\pi\rho_c \left[\frac{r^3}{3} - \frac{r^4}{4R_\star}  \right]$$

 The mass of the star, at $r = R_\star$ is:

 $$M_\star = 4\pi\rho_c \left[\frac{R_\star^3}{3} - \frac{R_\star^3}{4}  \right] =$$
 
  $$= 4\pi\rho_cR_\star^3 \left[\frac{1}{12} \right] = \frac{\pi\rho_cR_\star^3}{3} $$

  And so:

  $$\frac{M_r}{M_\star} = 4\pi\rho_c \left[\frac{r^3}{3} - \frac{r^4}{4R_\star}  \right] \left(\frac{3}{\pi\rho_cR_\star^3} \right) = 4 \left( \frac{r}{R_\star} \right)^3 - 3 \left(\frac{r}{R_\star} \right)^4$$

Hence:

$$\frac{g}{g_\star} = \frac{M_r}{M_\star} \left(\frac{R_\star}{r}\right)^2 = 4 \left( \frac{r}{R_\star} \right) - 3 \left(\frac{r}{R_\star} \right)^2$$

To find the maximum, we take the derivative of $\frac{g}{g_\star}$ with respect to $\frac{r}{R_\star}$, and set it equal to 0:

$$\frac{d (g/g_\star)}{d(r/R_\star)} = 4 - 6 \left(\frac{r}{R_\star} \right) = 0$$

Which implies, that the maximum occurs when:

$$\frac{r}{R_\star} = \frac{2}{3}$$

Finally, the value of $\frac{\rho}{\rho_\star}$ is obviously given by:

$$\frac{\rho}{\rho_c} = 1 - (r/R_\star)$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
# This option enables plots to be inline, 
# as opposed to opening in a separate window.

#The code is mostly from:
#https://github.com/veropetit/PHYS633-S2022/blob/main/L02-GravityField/02-GravityField-template.ipynb

fig, ax = plt.subplots(1,2, figsize=(11,6))
# The subplot routine creates a figure object (in the variable "fig"), which contains
# and an array (here 2 plots) of plotting windows called "axes" (in the variable "ax")

plt.rcParams.update({'font.size': 18})
plt.rcParams.update({'font.family': 'Times New Roman'})
# Set the font and font size for the whole figure.

for item in ax:
    item.set_xlabel(r'$r/R_\star$') # the "r" is to say there is latex formatted strings.
    item.set_xlim(0,1)
    item.set_ylabel(r'$\rho/\rho_c, M_r/M_\star, g/g_\star$') 
    item.set_ylim(0,1.5)
# ax is an vector of plot objects.
# we use a loop to set the title and limits of all plots to the same values

ax[0].set_title('Constant density')
ax[1].set_title(r'$\rho = \rho_c(1- r/R_\star)$')

#-----------------------------------------------------
#-----------------------------------------------------

#-----------------------------------------------------
# Part to do in class
#----

######################
#  Constant Density  #
######################

start = 0
stop = 1
num_stop = 100

r_rs = np.linspace(start, stop, num_stop)

M_Ms = (r_rs)**3
M_Ms[0] = 0

ax[0].plot(r_rs, M_Ms, label='Mass/Mstar')

g_gs = M_Ms * r_rs**2

ax[0].plot(r_rs, g_gs, label='gravity/surface_gravity', color='red')

########################
#  Decreasing Density  #
########################

rho_rhoc = 1.0 - r_rs
ax[1].plot(r_rs, rho_rhoc, label='density/central_density', color='red')

g_gs = (4*r_rs - 3*r_rs**2)

ax[1].plot(r_rs, g_gs, label='gravity/surface_gravity')

r_rs_max = 2/3

plt.axvline(x=r_rs_max, color='black') # I consulted the website: https://www.geeksforgeeks.org/plot-a-vertical-line-in-matplotlib/

#-----------------------------------------------------
# At home, add the curves for the g(s). 
# Analytically find at which r/Rstar is g at its maximum, and add a vertical line at this position.
#----





#-----------------------------------------------------
#-----------------------------------------------------
#------------------------------------------------------
for item in ax:
    item.legend(loc=0, fontsize=15)
plt.tight_layout()
# arrange the plot nicely

```
